```python
import os
import time
import uuid
```

Collecting images using cv2

```python
IMG_PATH = os.path.join('data','images')
num_img = 30
```

```python
import cv2
source = cv2.VideoCapture(0)
for imgnum in range(num_img):
    print('Colleting images {}'.format(imgnum))
    ret,frame = source.read()
    imgcode = os.path.join(IMG_PATH,f'{str(uuid.uuid1())}.jpg')
    cv2.imwrite(imgcode,frame)
    cv2.imshow('frames',frame)
    time.sleep(0.5)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

source.release()
cv2.destroyAllWindows()
```

Review Dataset and Build Image Loading Function

```python
import tensorflow as tf
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Avoid OOM errors by setting GPU Memory Consumption Growth
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus: 
    tf.config.experimental.set_memory_growth(gpu, True)
```

Load Image into TF Data Pipeline

```python
images = tf.data.Dataset.list_files('/home/spydy/joker/karm/face_detection_scratch/data/images/*.jpg')
```

```python
images.as_numpy_iterator().next()
```

```python
def load_image(x): 
    byte_img = tf.io.read_file(x)
    img = tf.io.decode_jpeg(byte_img)
    return img
```

```python
images = images.map(load_image)
```

```python
images.as_numpy_iterator().next()
```

```python
type(images)
```

 View Raw Images with Matplotlib

```python
image_generator = images.batch(5).as_numpy_iterator()
```

```python
plot_images = image_generator.next()
```

```python
fig, ax = plt.subplots(ncols=5, figsize=(20,20))
for idx, image in enumerate(plot_images):
    ax[idx].imshow(image) 
plt.show()
```

```python
for folder in ['train','test','val']:
    for file in os.listdir(os.path.join('data', folder, 'images')):
        
        filename = file.split('.')[0]+'.json'
        existing_filepath = os.path.join('data','label_img', filename)
        if os.path.exists(existing_filepath): 
            new_filepath = os.path.join('data',folder,'labels',filename)
            os.replace(existing_filepath, new_filepath)   
```
